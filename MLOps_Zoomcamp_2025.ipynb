```python
import pyarrow.parquet as pa
```

```python
jan_table = pa.read_table('yellow_tripdata_2023-02.parquet')
jan_table
```

```python
jan_df = jan_table.to_pandas()
```

```python
jan_df.head()
```

```python
jan_df['durations'] = (jan_df['tpep_dropoff_datetime']-jan_df['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
jan_df.head()
```

```python
durations_std = jan_df['durations'].std()
durations_std
```

```python
jan_df.count()
```

```python
jan_df = jan_df[(jan_df['durations']>=1)&(jan_df['durations']<=60)]
```

```python
train_jan_df=jan_df[['PULocationID','DOLocationID','durations']].copy()
```

```python
train_jan_df['PULocationID'] = train_jan_df['PULocationID'].astype(str)
train_jan_df['DOLocationID'] = train_jan_df['DOLocationID'].astype(str)
```

```python
train_jan_df.head()
```

```python
X_jan_data = train_jan_df[['PULocationID','DOLocationID']].to_dict('records')
```

```python
from sklearn.feature_extraction import DictVectorizer

X_vec_jan = DictVectorizer(sparse=True)

X_jan=X_vec_jan.fit_transform(X_jan_data)

feat_names = X_vec_jan.get_feature_names_out()
```

```python
Y_jan = train_jan_df['durations'].values
```

```python
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
jan_model = LinearRegression()
jan_model.fit(X_jan,Y_jan)
```

```python
y_jan_pred = jan_model.predict(X_jan)
```

```python
jan_rmse = np.sqrt(mean_squared_error(Y_jan,y_jan_pred))
jan_rmse
```

```python
feb_table = pa.read_table('yellow_tripdata_2023-02.parquet')
```

```python
feb_df = feb_table.to_pandas()
```

```python
feb_df['durations'] = (feb_df['tpep_dropoff_datetime']-feb_df['tpep_pickup_datetime']).dt.total_seconds() /60
```

```python
feb_df = feb_df[['PULocationID','DOLocationID','durations']]
```

```python
feb_df['PULocationID'] = feb_df['PULocationID'].astype(str)
feb_df['DOLocationID'] = feb_df['DOLocationID'].astype(str)
```

```python
feb_dict = feb_df[['PULocationID','DOLocationID']].to_dict('records')
```

```python
vec_feb = DictVectorizer(sparse=True)
X_feb = vec_feb.fit_transform(feb_dict)
```

```python
Y_feb = feb_df['durations'].values
```

```python
feb_model = LinearRegression()
feb_model.fit(X_feb,Y_feb)
```

```python
y_feb_pred = feb_model.predict(X_feb)
```

```python
rmse = np.sqrt(mean_squared_error(Y_feb,y_feb_pred))
rmse
```
